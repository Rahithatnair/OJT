# Use Iris dataset and train a decision tree model to predict the classes

```python
# Import the necessary libraries
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
```

```python

# Load the dataset
iris = load_iris()
X = iris.data[:, 2:] # petal length and width
y = iris.target
```

```python

# DecisionTreeClassifier
tree_clf = DecisionTreeClassifier(criterion='entropy',
								max_depth=2)
tree_clf.fit(X, y)
```

```python
from sklearn.tree import plot_tree
import numpy as np

# Assuming iris.target_names is a NumPy array
class_names = iris.target_names

# Convert the NumPy array to a list
class_names_list = class_names.tolist()

# Now use the list in the plot_tree function
plot_tree(tree_clf, feature_names=iris.feature_names,  
               class_names=class_names_list,
               filled=True)
```
